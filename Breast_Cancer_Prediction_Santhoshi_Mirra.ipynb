```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
cancer.describe()
```

```python
cancer.columns
```

```python
y = cancer['diagnosis']
```

```python
X = cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
labels = ["Benign", "Malignant"]
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)
```

```python
plt.title("Confusion Matrix - Breast Cancer Prediction (Logistic Regression)", fontsize=12)
```

```python
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
plt.figure(figsize=(4, 5))
plt.bar(["Logistic Regresison"], [accuracy], color="skyblue")
plt.ylim(0,1)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.text(0, accuracy + 0.01, f"{accuracy:2%}", ha='center', fontsize=10, fontweight='bold')
plt.show()
```
